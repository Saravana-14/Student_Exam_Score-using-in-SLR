# importing phase

```python
#importing the pandas libraries
import pandas as pd #Alice
```

```python
#read csv file for read pandas
df=pd.read_csv("student_exam_scores.csv")
```

```python
#print the variable for output
df
```

# Spliting process

```python
# it is using for seperate x and y
independent=df[["hours_studied"]]
```

```python
independent
```

```python
#same spliting process
dependent=df[["exam_score"]]
```

```python
dependent
```

# Train Test Split

```python
#splitting the dateset into training and testing
from sklearn.model_selection import train_test_split
```

```python
train_test_split(independent,dependent,test_size=0.2,random_state=0)
```

```python
# it is using for variables
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.2,random_state=0)
```

```python
x_train,x_test,y_train,y_test
```

```python
# it will generate the (how many rows & columns
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

# import the algorithm(train phase)

```python
# import the linear regression algorithm
from sklearn.linear_model import LinearRegression
LinearRegression()
```

```python
# variable
regressor=LinearRegression()
```

```python
regressor
```

```python
# fitting the model with training data
regressor.fit(x_train,y_train)
```

# To find weight and bias

```python
# to find the weight (slope)
regressor.coef_ #coefficient
```

```python
weight=regressor.coef_ #coefficient
```

```python
weight
```

```python
# this gives bias value
bias=regressor.intercept_
```

```python
bias
```

# Evalution metrices or finding parameters

```python
# predicting output for test data
y_predict=regressor.predict(x_test)
```

```python
# importing r2 score to evaluate model accuracy
from sklearn.metrics import r2_score
```

```python
# calculating  r2score
r2_score(y_test,y_predict)
```

```python
# variable
r_score=r2_score(y_test,y_predict)
```

```python
r_score
```

# save the model

```python
# importing the pickle for save the model
import pickle
```

```python
# file name
Filename="Final_model_SLR.save"
```

```python
# access the file
pickle.dump(regressor,open(Filename,'wb'))
```
